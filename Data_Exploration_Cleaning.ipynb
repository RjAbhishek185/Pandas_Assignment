```python
numeric_cols = df.select_dtypes(include=['number']).columns

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
text_cols = df.select_dtypes(include=['object']).columns

for col in text_cols:
    df[col] = df[col].fillna("Unknown")
```

```python
print(df.isnull().sum())
```

```python
before = len(df)

df.drop_duplicates(inplace=True)

after = len(df)

print("Duplicates Removed:", before - after)
```

```python
df[['title', 'rating', 'final_price', 'seller_name']].head()
```

```python
high_rated = df[df['rating'] > 4]

high_rated[['title', 'rating', 'final_price']].head()
```

```python
df['final_price'] = (
    df['final_price']
    .astype(str)
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.replace('"', '', regex=False)
)

df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce')
```

```python
df['quantity'] = 1

df['total_amount'] = df['final_price'] * df['quantity']
```

```python
df[['title', 'final_price', 'quantity', 'total_amount']].head()
```

```python
df.to_csv("cleaned_shopping_dataset.csv", index=False)

print("Cleaned Dataset Saved Successfully")
```
